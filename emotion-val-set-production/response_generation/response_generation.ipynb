```
import numpy as np
import csv
templates = {
    "grocery":["Okay, ordering "," to your address, is this correct?"],
    "calculator": ["Okay, the answer to your problem ", " Is this correct?"],
    "news": ["Reading the article from ", " Is this correct?"],
    "flight": ["I have booked your plane tickets to ", " Is this correct?"],
    "calendar": ["I am adding an event on ", " to your calendar. Is this correct?"],
    "message": ["Alright, your message says ", "Is this correct?"],
    "call": ["Okay, calling ", " Is this correct?"],
    "music": ["Okay, playing the ", " . Is this correct?"],
    "timer": ["Okay, setting a timer for ", " minutes. Is this correct?"]
}

def parse(topic):
  data = open("drive/MyDrive/Voice_Assistant_Data/" + topic + ".txt", 'r', encoding='utf-8')
  responses = []
  for line in data:
    response = templates[topic][0] + line[:-1] + templates[topic][1]
    responses.append(response)
  return responses

with open('drive/MyDrive/Voice_Assistant_Data/prompt_to_response.csv', 'w', newline='') as csvfile:
  writer = csv.writer(csvfile, delimiter=',', quotechar='"')
  for key in templates:
    ## NLP Error
    responses = parse(key)
    for response in responses:
      writer.writerow(["prompts/" + key + ".png", response])
  
    ## Failed Feedback
    writer.writerow(["prompts/" + key + ".png", "I'm sorry, I cannot do that"])
    writer.writerow(["prompts/" + key + ".png", "I'm sorry, I don't know how to do that"])

    ## Unfamiliar Intent or System Error
    writer.writerow(["prompts/" + key + ".png", "I'm sorry, I didn't catch that"])
```
